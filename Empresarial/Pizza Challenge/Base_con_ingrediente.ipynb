### Transformacion Base: se agrego a la base integrada con SQL Server las columnas de: 'ingredient' e 'ingredient_code".

Esta base se utilizara para el Informe Operacional para gestionar el inventario de ingredientes por tipo de pizza.

```python
!pip install pandas
import pandas as pd
```

```python
# Leer las Tablas
df_base = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Base.xlsx")
df_pizza_ingredients = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/pizza_ingredients.xlsx')
```

```python
# Agregar las columnas ingredient e ingredient_code a la tabla Base
df_base = pd.merge(df_base, df_pizza_ingredients[['pizza_id', 'ingredient', 'ingredient_code']], on='pizza_id', how='left')
df_base
```

```python
# Guardar la tabla Base con las nuevas columnas en un archivo CSV
df_base.to_csv('Base_con_ingredientes.csv', index=False)
```
